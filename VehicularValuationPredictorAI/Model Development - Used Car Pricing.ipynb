# Trying Polynomial Regression and Pipelines

```python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')

    plt.show()
    plt.close()
```

```python
x=df['highway-mpg']
y=df['price']
```

```python
#using a polynomial of the 3rd order (cubic) 
f= np.polyfit(x,y,3)
p= np.poly1d(f)
```

```python
PlotPolly(p,x,y,'highway-mpg')
```

```python
F= np.polyfit(x,y,11)
P= np.poly1d(F)
```

```python
PlotPolly(P,x,y,'highway mpg')
```

```python
m= np.polyfit(x,y,14)
s= np.poly1d(m)
PlotPolly(s,x,y,'highway mpg')
#if we put order =15 the fitting turns to be inaccurate
```


the expression for a second-order (degree=2) polynomial with two variables is given by:
$$
Yhat = a + b_1 X_1 +b_2 X_2 +b_3 X_1 X_2+b_4 X_1^2+b_5 X_2^2
$$

Creating Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
Input = [('scaler', StandardScaler()),('polynomial', PolynomialFeatures(interaction_only=False)),('model',LinearRegression())]
```

```python
pipe = Pipeline(Input)
pipe
```

convert the data type Z to type float to avoid conversion warnings that may appear as a result of StandardScaler taking float inputs

```python
Z = Z.astype(float)
pipe.fit(Z,y)
```

```python
ypipe=pipe.predict(Z)
ypipe[0:4]
```

trying a pipeline that standardizes the data, then produce a prediction using a linear regression model using the features Z and target y.

```python
Input = [('scale', StandardScaler()),('model',LinearRegression())]
pipe = Pipeline(Input)
pipe.fit(Z,y)

ypipe=pipe.predict(Z)
ypipe[0:10]
```

#  Getting info of Measures for In-Sample Evaluation
for accuracy

Model 1: Simple Linear Regression

```python
#highway_mpg_fit
x= df[['highway-mpg']]
y= df[['price']]
lm.fit(x, y)
# Find the R^2
print('The R-square is: ', lm.score(x, y))
```

```python
Yhat=lm.predict(x)
print('The output of the first four predicted value is: ', Yhat[0:4])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse= mean_squared_error(df['price'], Yhat)
print('The mean square error of price and predicted value is: ', mse)
```
